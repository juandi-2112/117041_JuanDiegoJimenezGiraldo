```python
#Se importa la libreria keyword, que es donde están las palabras reservadas
import keyword
#Se ingresa una cadena de texto 
g = input('Ingrese un texto: ')

#Se define la cadena de vocales mayúsculas
vcl = 'AEIOU'
#Se define la cadena de vocales con tildes
vcl_til = 'áéíóú'
#Se define la cadena de vocales con tildes y en mayúsucula
vcl_tilmay = 'ÁÉÍÓÚ'
#Se define la cadena de espacio
bac = ' '
#Se define resr que es la lista con las palabras reservadas
resr = (keyword.kwlist)
#Se definen los contadores vacios, para allí alojar los datos que se requieran
#A paritr de la cadena ingresada por el usuario
c_vcl, c_vcl_til, c_resr, c_bac, c_gt = 0, 0, 0, 0, 0
#Se hace que el texto ingresado se convierta en una cadena
lst = g.split(' ')  
#Aquí se hacen las iteraciones para determinar que condiciones se cumple
#por ejemplo si existen digitos en el texto ingresado, y si cumple alguna 
#o algunas de las condiciones estas se iran alojando en los contadores
#que les corresponda
for i in range(len(g)):
    if g[i].isdigit() is True:
        c_gt += 1
    if g[i] in bac:
        c_bac += 1
    if g[i] in vcl_tilmay or g[i] in vcl_til:
        c_vcl_til += 1
    if g[i] in vcl or g[i] in vcl_tilmay:
        c_vcl += 1
#Esto es para determinar si en texto ingresado existe una plabnra reservada,
#y en el caso de que exista, la agrega en el contador específico
#para ello
for m in lst:
    if m in resr:
        c_resr += 1
#Imprime los contadores mencionados anteriormente, para que de esta manera
#se pueda derterminar cuales condiciones se cumplen y cuantas veces estas 
#ocurren en el texto ingresado        
print(f'El número de vocales mayúsculas que se entraron es: {c_vcl} vocales\n'
f'El número de letras con tildes que se entaron es de: {c_vcl_til} letras\n'
f'El número de dígitos que se entraron es: {c_gt} dígitos\n'
f'El número de espacios que se entraron es: {c_bac} espacios\n'
f'El número de palabras resevadas que se entraron es: {c_resr} palabras')
    
```

```python
#Esta funcion se define con el fin de que el usuario entre correctamente
#una matriz, para poder realizarle las operaciones necesarias a la misma
def readmx():
    
#Se utiliza el try - except para que la matriz ingresada este compuesta
#de números enteros, y esta sea correcta, ya que de esta forma, si se podrá
#operar a la misma
    while True:
        try:
            mx= eval(input('Ingrese una matriz cuadrada: '))
            if type(mx) is list:
                for ro in mx:
                    if type(ro) is list:
                        for nti in ro:
                            if type(nti) is not int:
                                raise Exception
                    else:
                        raise Exception
            else:
                raise Exception
            
            for rw in mx:
                if len(rw) != len(mx[0]) :
                    raise Exception
            break     
        except:
            print('Se ha ingresado una matriz inválida')
#Se terorna la matriz 'mx' para poder trabajar con la misma    
    return mx

#Está función se hace con el fin de que se muestre al usuario la matriz 
#de la forma que se está acostumbrado
def prtmx(mx):
    for rw in mx:
        for col in rw:
            print('{:8}'.format(col),end=' ')
        print()

#Con la siguinte iteración es posible determinar si la matriz ingresada por 
#el usuario es cuadrada
        
while True:    
    mx_a = readmx()
    if len(mx_a[0])==len(mx_a):
        break
    else:
        print('La matriz ingresada no es válida. Recuerde que esta debe ser '
              'cuadrada -número de filas igual al número de columnas-')
print()
print('La matriz A es: ')

#Esto configura la matriz en su forma convencional
prtmx(mx_a)

#Con esta iteración es posible triangular la matriz a mx_a 
for ro in range(0,len(mx_a)):
    for col in range(0,len(mx_a[0])):
        if ro== col :
            break
        else :
            mx_a[ro][col]=0
print()
            
print('La matriz asociada U es: ')

prtmx(mx_a)

#Se crea una lista vacía donde se alojarán unos datos necesarios 
lst=[]

#Aquí se cuenta la cantidad de unos que estan en la matriz
for ro in mx_a:
    g = 0
    for nti in ro:
        if nti == 1:
            g += 1
    lst.append(g)
    
#Se define b como el contador donde se encuentra la mayor cantidad de unos
b = max(lst)

#Se define v para poder identificra en que lista se encuentra la mayor cantidad
#de unos
v = lst.index(b)

#Se imprime el número de la fila, la cual tiene la mayor cantidad de unos
if b == 0:
    print('No hay "unos" en la matriz')
else:
    print(f'la fila con mayor cantidad de unos es {v+1}')
```

```python
#Se importa la libreria numpy ya que de esta se utilizará el comando rot
#que es el que permite rotar la matriz
import numpy as np


def readmx():
#Se utiliza el try - except para que la matriz ingresada este compuesta
#de números enteros, y esta sea correcta, ya que de esta forma, si se podrá
#operar a la misma
     while True:
        try:
            mx= eval(input('Ingrese una matriz: '))
            if type(mx) is list:
                for ro in mx:
                    if type(ro) is list:
                        for nti in ro:
                            if type(nti) is not int:
                                raise Exception
                    else:
                        raise Exception
            else:
                raise Exception
            
            for rw in mx:
                if len(rw) != len(mx[0]) :
                    raise Exception
            break     
        except:
            print('Se ha ingresado una matriz inválida')
#Se terorna la matriz 'mx' para poder trabajar con la misma    
     return mx
#Está función se hace con el fin de que se muestre al usuario la matriz 
#de la forma que se está acostumbrado 
def prtmx(mx):
    for rw in mx:
        for col in rw:
            print('{:8}'.format(col),end=' ')
        print()
        
     
#Se define la matriz mx_a
mx_a = readmx()
 
print()
print('La matriz A es: ')

prtmx(mx_a)

#Aquí lo que hace el rot90 es girar la matriz 90 grados en sentido 
#antihorario
w = np.rot90(mx_a)

print()

print('La matriz rotada 90 grados sería: ')
#Se imprime la nueva matriz ya rotada 90 grados en sentido antihorario 
prtmx(w)
```
